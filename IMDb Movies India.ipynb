```python
# cloumns with lest missing values 
cols = [var for var in Movie_data_set.columns if Movie_data_set[var].isnull().mean() < 0.21 and Movie_data_set[var].isnull().mean() > 0]
Movie_data_set.dropna(subset=cols , inplace = True)
```

```python
Movie_data_set.head()
```

```python
Movie_data_set['Year'] = Movie_data_set['Year'].str.extract('(\d+)').astype('int64')
Movie_data_set['Duration'] = Movie_data_set['Duration'].fillna('0').str.extract('(\d+)').astype('int64')
mean_value_Duration=round(Movie_data_set['Duration'].mean(),0)
Movie_data_set['Duration']=Movie_data_set['Duration'].replace(0,mean_value_Duration)
```

```python
Movie_data_set.sample(50)
```

```python

Movie_data_set["Votes"] = Movie_data_set["Votes"].str.replace('[\$,M]', '', regex=True)
Movie_data_set["Votes"] = Movie_data_set["Votes"].fillna('0').astype(float).astype(int)
mean_value_Votes=round(Movie_data_set['Votes'].mean(),0)
Movie_data_set['Votes']=Movie_data_set['Votes'].replace(0,mean_value_Votes)
```

```python
categorized_clolumn_director=Movie_data_set_1["Director"].value_counts()
categorized_clolumn_Actor_1=Movie_data_set_1["Actor 1"].value_counts()
categorized_clolumn_Actor_2=Movie_data_set_1["Actor 2"].value_counts()
categorized_clolumn_Actor_3=Movie_data_set_1["Actor 3"].value_counts()

Movie_data_set_1["categorized_Director"]=Movie_data_set_1["Director"].map(categorized_clolumn_director)
Movie_data_set_1["categorized_Actor_1"]=Movie_data_set_1["Actor 1"].map(categorized_clolumn_Actor_1)
Movie_data_set_1["categorized_Actor_2"]=Movie_data_set_1["Actor 2"].map(categorized_clolumn_Actor_2)
Movie_data_set_1["categorized_Actor_3"]=Movie_data_set_1["Actor 3"].map(categorized_clolumn_Actor_3)
```

```python
Movie_data_for_model_traning=Movie_data_set_1[Movie_data_set_1. Rating.notnull()]
Movie_data_for_model_testing=Movie_data_set_1[Movie_data_set_1. Rating.isnull()]
```

```python
x_train = Movie_data_for_model_traning.drop("Rating",axis=1)
y_train = Movie_data_for_model_traning["Rating"]

x_test = Movie_data_for_model_testing.drop("Rating",axis=1)
y_test = Movie_data_for_model_testing["Rating"]
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()

# Train the classifier on the training data
linear_model.fit(x_train, y_train)

# Make predictions on the testing data
y_predection = linear_model.predict(x_test)
```

```python
Movie_data_for_model_testing["Rating"]=y_predection 
Movie_data_for_model_testing.head()
```

```python
completed_data_for_training_model = pd.concat([Movie_data_for_model_traning,Movie_data_for_model_testing], axis=0)
completed_data_for_training_model['Rating'] = completed_data_for_training_model['Rating'].round(0)
completed_data_for_training_model
```

```python

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Generate sample data (replace with your own dataset)
np.random.seed(0)
X = completed_data_for_training_model.drop("Rating", axis=1)
Y = completed_data_for_training_model["Rating"]

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# Create a Linear Regression model
linear_model = LinearRegression()

# Train the model on the training data
linear_model.fit(X_train, Y_train)

# Make predictions on the testing data
Y_prediction = linear_model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(Y_test, Y_prediction)
r2 = r2_score(Y_test, Y_prediction)

# Print the results
print('Mean Squared Error:', mse)
print('R-squared:', r2)
```
